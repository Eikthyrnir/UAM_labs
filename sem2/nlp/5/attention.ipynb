```python
import torch
import math
from torch import bmm
```

```python
def step_by_step_batch_attention(Q, K, V):
    d_k = Q.size(-1)
    scores = bmm(Q, K.mT) / math.sqrt(d_k)
    print(f"scores = \n{scores}")
    score_probabilities = torch.softmax(scores, dim=-1)
    print(f"softmax(scores) = \n{score_probabilities}")
    assert torch.all((torch.sum(score_probabilities, dim=-1) - 1.0).abs() < 1e-6), "Softmax rows do not sum to 1"
    return bmm(score_probabilities, V)
```

```python
# Sample values are small so that it's easier to check correctness of the calculations
n = 4 # words count
d = 2 # size of embedding
d_k = 3
batch_size = 2

X = torch.rand(n, d)
W_q = torch.rand(batch_size, d, d_k)
W_k = torch.rand(batch_size, d, d_k)
W_v = torch.rand(batch_size, d, d_k)
```

```python
print(f"X = \n{X}")
print(f"W_q = \n{W_q}")
print(f"W_k = \n{W_k}")
print(f"W_v = \n{W_v}")
```

```python
X_batch = X.unsqueeze(0).expand(batch_size, -1, -1)
Q = torch.bmm(X_batch, W_q)
K = torch.bmm(X_batch, W_k)
V = torch.bmm(X_batch, W_v)

print(f"Q = \n{Q}")
print(f"K = \n{K}")
print(f"V = \n{V}")
```

```python
attention_results = step_by_step_batch_attention(Q, K, V)
```

```python
print(f"attention = \n{attention_results}")
```
