
## **Zadania: Ewolucja i kompromisy w detekcji obiektów: R-CNN, Faster R-CNN oraz SSD**

**Cel zadania:**

Zrozumienie fundamentalnych różnic architektonicznych między detektorami dwuetapowymi a jednoetapowymi oraz samodzielna analiza kompromisu (trade-off) między precyzją (mAP) a czasem wnioskowania (FPS).

#### **Wymagane elementy do oddania:**

* Raport (np. w formacie PDF) zawierający wypełnioną tabelę porównawczą.
* Uzasadnienia dla wszystkich trzech scenariuszy z Części 2.
* Kod źródłowy skryptu testującego, wygenerowane wykresy FPS oraz zestawienie wizualne (screenshoty) obrazujące różnice w predykcjach


#### **Część 1: Architektoniczna analiza porównawcza (Teoria)**

Zadaniem jest opracowanie szczegółowej tabeli porównawczej dla modeli: **R-CNN**, **Faster R-CNN** oraz **SSD**. Tabela powinna zawierać następujące kryteria:

* **Metoda generowania propozycji regionów:** (np. Selective Search, RPN, domyślne ramki/kotwice).
* **Podejście do treningu:** Czy sieć uczy się w pełni *end-to-end*, czy wymaga wieloetapowego trenowania osobnych modułów?
* **Sposób radzenia sobie z wieloma skalami obiektów:** (np. zmiana rozmiaru obrazu wejściowego, ekstrakcja cech z wielu warstw).
* **Szybkość wnioskowania (FPS):** Rząd wielkości dla standardowego sprzętu.
* **Wąskie gardło (Bottleneck):** Co najbardziej spowalnia dany algorytm lub sprawia największe problemy podczas uczenia?


#### **Część 2: Studium przypadku – Wybór architektury (Analiza)**

Rozważ trzy scenariusze projektowe. Dla każdego z nich wybierz **jeden** z omawianych algorytmów, który najlepiej sprawdzi się w danym zadaniu, oraz szczegółowo uzasadnij swój wybór, powołując się na cechy z Części 1.

* **Scenariusz A: System unikania kolizji w dronie.** Dron porusza się z dużą prędkością i ma ograniczoną moc obliczeniową (procesor brzegowy/edge device). Musi wykrywać zbliżające się przeszkody (ptaki, drzewa, inne drony).
* **Scenariusz B: Analiza zdjęć histopatologicznych (medycznych).** System działa na potężnym serwerze w szpitalu. Jego celem jest znalezienie bardzo rzadkich i niezwykle małych komórek nowotworowych na obrazach o gigantycznej rozdzielczości. Czas analizy pojedynczego zdjęcia nie ma większego znaczenia, liczy się bezbłędność.
* **Scenariusz C: Historyczna analiza techniczna.** Wyobraź sobie, że jest rok 2014, a Ty musisz stworzyć system do tagowania zdjęć użytkowników w aplikacji społecznościowej (rozpoznawanie twarzy, zwierząt, samochodów). Nie masz dostępu do sieci *end-to-end*, ale zależy Ci na lepszej dokładności niż proste *sliding window*.

#### **Część 3: Eksperyment empiryczny**


Pobierz gotowe, pretrenowane wagi modeli **Faster R-CNN** oraz **SSD** (np. z biblioteki torchvision.models w PyTorch lub TensorFlow Hub).

1. **Zbiór testowy:** Uruchom oba modele na krótkim pliku wideo (np. 30-sekundowy ruch uliczny) oraz na 5 statycznych zdjęciach o dużym zagęszczeniu małych obiektów.
2. **Pomiary:** Zmierz średni czas przetwarzania jednej klatki (w milisekundach) i przeliczyć go na FPS dla obu modeli.
3. **Wnioski wizualne:** Wskazać na wygenerowanych obrazach przykłady sytuacji, w których Faster R-CNN poradził sobie lepiej od SSD (np. z wykryciem małego obiektu w tle) i odwrotnie.
